```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('ecg.csv')


```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

X_train, X_test = train_test_split(data_scaled, test_size=0.2, random_state=42)
```

```python
input_dim = X_train.shape[1]
encoding_dim = 32

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)

encoder = Model(input_layer, encoded)
```

```python
decoded = Dense(input_dim, activation='sigmoid')(encoded)

autoencoder = Model(input_layer, decoded)
```

```python
autoencoder.compile(optimizer=Adam(), loss='mean_squared_error')
```

```python
history = autoencoder.fit(X_train, X_train, epochs=50, batch_size=256, validation_data=(X_test, X_test))
```

```python
X_train_pred = autoencoder.predict(X_train)
X_test_pred = autoencoder.predict(X_test)

train_mse = np.mean(np.power(X_train - X_train_pred, 2), axis=1)
test_mse = np.mean(np.power(X_test - X_test_pred, 2), axis=1)

threshold = np.percentile(train_mse, 95)

anomalies = test_mse > threshold
print(f"Number of anomalies detected: {np.sum(anomalies)}")

plt.hist(test_mse, bins=50)
plt.axvline(threshold, color='r', linestyle='--')
plt.title("Reconstruction Error Distribution")
plt.show()
```

```python
for i in range(len(X_test)):
    if anomalies[i]:
        plt.plot(X_test[i])
        plt.title(f"Anomaly Detected: Sample {i}")
        plt.show()
```
